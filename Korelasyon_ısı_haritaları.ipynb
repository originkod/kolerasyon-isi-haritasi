```python
import pandas as pd #import ile kullanacağımız kütüphaneyi yükleyip "as" ile yüklediğimiz kütüphaneye isim veriyoruz.
import seaborn as sns
import matplotlib.pyplot as plt

# Iris veri setini yükleyin (Pandas ile bulunan bir örnek veri setidir)
data = sns.load_dataset("iris") # load_dataset fonksiyonu, Seaborn tarafından sağlanan içinde bulunan örnek veri setlerini yüklemek ve kullanmak için kullanılır.
# Bağımsız değişkenleri seçin
X = data.iloc[:, :-1] #iloc, "integer location"ın kısaltmasıdır. Bu özellik, veri çerçevelerinde (DataFrame) belirli satır ve sütunlara erişmek için kullanılır. iloc, konumları belirlemek için tamsayı tabanlı bir endeksleme yöntemidir.

# Hedef değişkeni seçin
y = data.iloc[:, -1]

# Veri setindeki her özelliğin korelasyonunu alın
correlation_matrix = data.corr() #corr() fonksiyonu, "data" veri seti üzerinde tüm sütunlar arasındaki korelasyonları hesaplar.

# En yüksek korelasyona sahip özellikleri seçin
top_corr_features = correlation_matrix.index

# Isı haritasını çizmek için bir figür oluşturun
plt.figure(figsize=(10, 8))

# Korelasyon ısı haritasını çizin
g = sns.heatmap(data[top_corr_features].corr(),
                annot=True, cmap="RdYlGn")

# Isı haritasını göster
plt.show()
```

Bu, Iris veri kümesindeki en önemli 2 özelliği ve bu özelliklerin skorlarını gösterir.
